```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
# Dataset Reading
df_salary = pd.read_csv('Dataset/Salary_Data.csv')
```

```python
df_salary.head()
```

```python
# Dataset Description
df_salary.describe()
```

```python
# Dataset Distribution
plt.title('Salary Distribution')
sns.histplot(df_salary.Salary,kde=True, stat="density", linewidth=1)
plt.show()
```

```python
#Relationships between the salary and years of experience variables

plt.scatter(df_salary.YearsExperience, df_salary.Salary, color = 'Red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Splitting variables for X and Y learning
X = df_salary.iloc[:, :-1]
y = df_salary.iloc[:, 1:]
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Create the model and pass the data
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Prediction
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluate performance on training set
mae_train = mean_absolute_error(y_train, y_pred_train)
mse_train = mean_squared_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

print("Training set evaluation:")
print(f"Mean Absolute Error: {mae_train}")
print(f"Mean Squared Error: {mse_train}")
print(f"R-squared Score: {r2_train}")
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
